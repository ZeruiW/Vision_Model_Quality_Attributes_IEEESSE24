```python
# 计算每个模型的解释弹性得分
explanation_resilience_scores = {}

for model in xairs:
    original_scores = np.array(list(map(float, xairs[model]['Original'].values())))
    perturbation_changes = []

    for perturbation in xairs[model]:
        if perturbation != 'Original':
            perturbation_scores = np.array(list(map(float, xairs[model][perturbation].values())))
            change = np.mean(perturbation_scores - original_scores)
            perturbation_changes.append(change)

    # 计算平均变化，并应用1-x转换
    avg_change = np.mean(perturbation_changes)
    resilience_score = 1 - avg_change

    # 考虑原始可解释性得分：如果原始得分较差（值较大），则降低解释弹性得分
    original_avg_score = np.mean(original_scores)
    adjusted_resilience_score = resilience_score * (1 - original_avg_score)
    
    # 确保分数在0到1之间
    adjusted_resilience_score = min(max(adjusted_resilience_score, 0), 1)

    explanation_resilience_scores[model] = adjusted_resilience_score

# 创建一个新的DataFrame来保存解释弹性得分
df_explanation_resilience = pd.DataFrame.from_dict(explanation_resilience_scores, orient='index', columns=['Explanation Resilience'])

df_explanation_resilience
```
